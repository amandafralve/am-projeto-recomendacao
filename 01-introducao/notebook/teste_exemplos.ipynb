```python
import matplotlib.pyplot as plt

valores = [1, 2, 3, 4, 5]
plt.plot(valores)
plt.title("Gráfico simples")
plt.show()
```

```python
import matplotlib.pyplot as plt

categorias = ["A", "B", "C"]
valores = [10, 5, 8]

plt.bar(categorias, valores)
plt.title("Gráfico de barras")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x, y)
plt.title("Gráfico de dispersão")
plt.show()
```
