```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import numpy as np 
import time
```

```python
## Solução com Vetores
# Vetor de Entradas [energy, loudness_norm]
X = [0.9, 0.6]

# Vetor de Pesos [peso_energy, peso_loudness]
W = [0.8, 0.2]

z_manual = (X[0] * W[0] + X[1] * W[1])
print(f"Cálculo Manual z = {z_manual} ")
```

```python
# Solução com Numpy 
X_np = np.array(X)
W_np = np.array(W)

z_numpy = np.dot(X_np, W_np)

print(f"Cálculo Numpy: z = {z_numpy}")
print(f"\n Resultados : {z_manual} = {z_numpy}")
```
