```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('../data/raw/dataset.csv')
df.head()
df.describe()
```

```python
plt.figure(figsize=(10,6))

plt.hist(df['danceability'], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.xlabel('Danceability')
plt.ylabel('Frequencia')
plt.title('Distribuicao de dancabilidade')
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.hist(df['energy'], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.xlabel('Energy')
plt.ylabel('Frequencia')
plt.title('Distribuicao de energia')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['energy'], df['loudness'], alpha=0.5, edgecolors='black')
plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.title('Energia vs Loudness')
plt.show()
```

```python
# Filtrando exceções: Alta energia e Baixo volume
excecoes = df[(df['energy']> 0.8) & (df['loudness'] < -10)]

# Ordenando pelas mais energéticas
excecoes_ordenadas = excecoes[['track_id','track_name','artists','energy','loudness']].sort_values(by='energy', ascending=False)

# Exibindo as top 10
excecoes_ordenadas.head(10)
```
