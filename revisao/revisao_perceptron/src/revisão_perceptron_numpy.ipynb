```python
import numpy as np 

# O vetor [ Renda, Historico, Idade, Garantia]
pesos = np.array([0.3, 0.8, 0.1, 0.5])

bias = -9.0
```

```python
# Cliente teste
# [ Renda, Historico, Idade, Garantia]
cliente_teste = np.array([7.0, 8.0, 4.0, 6.0])

# Cálculo Z (Equação de Decisão)
z = np.dot(cliente_teste, pesos) + bias

# Função de ativação (Degrau/Step)
decisao = 1 if z >= 0 else 0
print(f"O valor de Z foi {z:.2f} | Decisão : {'Aprovado' if decisao == 1 else 'Reprovado'}")
```

```python
# Cálculo em lote (vários clientes de uma vez)
clientes_lote = np.array([
    [ 7.0, 8.0 4.0, 6.0 ], # Cliente A (Bom perfil geral)
    [ 3.0, 2.0, 5.0, 0.0 ], # Cliente B (Risco alto, sem garantia)
    [ 5.0, 5.0, 5.0, 10.0 ], # Cliente C (Renda média, mas com ótima garantia)
    # Cliente D tentando APROVAR alterando a renda
    [ 0.0, 0.0, 5.0, 0.0 ]
])

Z_lote = np.dot(clientes_lote * pesos) + bias

decisao_lote = (Z_lote >= 0).astype(int)

# Saída
for i, decisao in enumerate(decisao_lote):
    status = "Aprovado" if decisao == 1 else "Reprovado"
    print(f"Cliente {i+1} -> Z = {Z_lote[i]:.2f} | Veredito: {status} )
```

```python
from perceptron_numpy import PerceptronNumpy

nosso_modelo = PerceptronNumpy(pesos=[0.3, 0.8, 0.1, 0.5], bias=-9.0)

resultado = nosso_modelo.predict(clientes_lote)

print("Status completo da tabela de lote:", resultado)
```
