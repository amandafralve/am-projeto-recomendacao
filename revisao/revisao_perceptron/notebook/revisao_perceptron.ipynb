```python
# Entradas
renda_mensal = 6.0 # o a 10 representando faixas salariais
historico_pagador = 5.0 # 0 a 10 Bom pagador
```

```python
# Pesos
peso_renda = 0.3
peso_historico = 0.8

# Bias (A resistência natural em emprestar dinheiro)
# Bias positivo demais entende a dizer sim para qualquer situação
bias = -7.0
```

```python
# Cálculo do Z (Equação de decisão)
z = (renda_mensal + peso_renda) + (historico_pagador * peso_historico) + bias

print(f"O valor interno gerado pelo Perceptron = {z:.2f})
```

```python
# Função de Ativação (Degrau / Step)
if z >=0:
    decisao = 1
    motivo = "Aprovado! Bom pagador."
else:
    decisao = 0
    motivo = "Reporovado. Risco muito alto"

print(f"Decisão Final {decisao}")
print(f"Motivo {motivo}")
```
